```python
import pandas as pd
import joblib
from pathlib import Path

PROJECT_ROOT = Path(r"C:\Users\Anna\PycharmProjects\churn-mlops-telco")

# Dataset brut (avec customerID)
df_raw = pd.read_csv(PROJECT_ROOT / "data" / "raw" / "WA_Fn-UseC_-Telco-Customer-Churn.csv")

# X_test (sans customerID)
X_test = joblib.load(PROJECT_ROOT / "data" / "processed" / "X_test.joblib")

print(df_raw.shape)
print(X_test.shape)
```

```python
X_test.index[:5]
df_raw.loc[X_test.index].head()
```

```python
# On récupère customerID pour les mêmes lignes que X_test
df_test_with_id = X_test.copy()
df_test_with_id.insert(
    0,  # première colonne
    "customerID",
    df_raw.loc[X_test.index, "customerID"].values
)

df_test_with_id.head()
```

```python
csv_path = PROJECT_ROOT / "sample_clients_with_id.csv"
df_test_with_id.head(50).to_csv(csv_path, index=False)

csv_path
```
